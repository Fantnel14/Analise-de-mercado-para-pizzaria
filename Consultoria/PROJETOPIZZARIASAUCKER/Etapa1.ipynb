```python
# Importação das bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Estilo dos gráficos para ficar mais bonito
sns.set(style="whitegrid")
```

```python
# Carregando o dataset (substitua pelo caminho do seu arquivo)
df1 = pd.read_excel(r'C:\Users\mathe\Dropbox\Consultoria\data\2022.09.22_Síntese-Estatística-do-DF.xls', sheet_name=1 )
df2 = pd.read_excel(r'C:\Users\mathe\Dropbox\Consultoria\data\16_LAGO-SUL.xlsx')
```

```python
# Filtrando o DataFrame para remover o Distrito Federal
df1_filtrado = df1[df1['Nome'] != 'Distrito Federal²']

plt.figure(figsize=(12, 8))

# Gráfico de dispersão (População x Renda per Capita)
plt.scatter(df1_filtrado['Renda domiciliar per capita³ (R$)'], df1_filtrado['População Total'], alpha=0.7, color='blue')

# Rótulos para cada ponto (Região)
for i, txt in enumerate(df1_filtrado['Nome']):
    plt.annotate(txt, (df1_filtrado['Renda domiciliar per capita³ (R$)'].iloc[i], 
                        df1_filtrado['População Total'].iloc[i]), fontsize=9, alpha=0.7)

# Adicionando título e rótulos aos eixos
plt.title('Relação entre População e Renda Per Capita por Região', fontsize=16)
plt.xlabel('Renda Per Capita (R$)', fontsize=12)
plt.ylabel('População Total', fontsize=12)

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

```python

# Calcular a média de renda per capita e população total do DF
media_renda_df = df1_filtrado['Renda domiciliar per capita³ (R$)'].mean()
media_populacao_df = df1_filtrado['População Total'].mean()

# Obter dados do Lago Sul
lago_sul = df1_filtrado[df1_filtrado['Nome'].str.strip().str.lower() == 'lago sul']
renda_lago_sul = lago_sul['Renda domiciliar per capita³ (R$)'].values[0]
populacao_lago_sul = lago_sul['População Total'].values[0]

# Gráfico 1: Comparativo da Renda Per Capita
plt.figure(figsize=(10, 6))

plt.bar(['Média do DF', 'Lago Sul'], [media_renda_df, renda_lago_sul], color=['blue', 'green'])
plt.title('Comparativo de Renda Domiciliar Per Capita: Lago Sul vs Média do DF', fontsize=14)
plt.ylabel('Renda Per Capita (R$)', fontsize=12)

# Destacar o quanto o Lago Sul está acima da média
for i, valor in enumerate([media_renda_df, renda_lago_sul]):
    plt.text(i, valor + 2000, f'R${valor:,.2f}', ha='center', fontsize=10)  # Aumentar o deslocamento vertical

plt.ylim(0, renda_lago_sul + 3000)  # Ajuste do eixo Y para criar mais espaço acima das barras
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```
